```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score

path_of_file = r"Lab Session Data.xlsx"
xl = pd.ExcelFile(path_of_file)
# MSE, RMSE, MAPE and R2 scores for the price prediction exercise done in Lab 02.
price = pd.read_excel(xl, sheet_name = 'IRCTC Stock Price')
price['Predicted_Price'] = price['Price'].shift(1)
price = price.dropna()

real_price = price['Price'].values
assumed_price = price['Predicted_Price'].values

mse = mean_squared_error(real_price, assumed_price)
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(real_price, assumed_price)
r2 = r2_score(real_price, assumed_price)

print("mean squared error is:",mse)
print("root mean squared error is:",rmse)
print("mean absolute percentage error is:",mape)
print("r2 score is:",r2)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Load your dataset
df = pd.read_excel('training_mathbert 3 test.xlsx')  # Update the file name/path as needed

# Subset the data for these two features and the target class label
X = df[['embed_0', 'embed_1']]
y = df['output']

# Exercise A3: Plotting the training data
plt.figure(figsize=(10, 6))
colors = {0: 'blue', 1: 'red'}
plt.scatter(X['embed_0'], X['embed_1'], c=y.apply(lambda label: colors[label]), label='Training Data')
plt.xlabel('embed_0')
plt.ylabel('embed_1')
plt.title('Training Data Scatter Plot')
plt.legend()
plt.show()

# Exercise A4: Generate test data and classify
# Generating a grid of test points
x_test = np.arange(X['embed_0'].min(), X['embed_0'].max(), 0.1)
y_test = np.arange(X['embed_1'].min(), X['embed_1'].max(), 0.1)
xx, yy = np.meshgrid(x_test, y_test)
X_test = np.c_[xx.ravel(), yy.ravel()]

# Convert the test data to a DataFrame with the same feature names
X_test_df = pd.DataFrame(X_test, columns=['embed_0', 'embed_1'])

# kNN classification with k=3
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)
pred = knn.predict(X_test_df)

# Plot test data predictions
plt.figure(figsize=(10, 6))
plt.scatter(X_test_df['embed_0'], X_test_df['embed_1'], c=pred, alpha=0.5)
plt.xlabel('embed_0')
plt.ylabel('embed_1')
plt.title('Test Data Classification with k=3')
plt.show()

# Exercise A5: Repeat classification for different k values
for k in [1, 5, 7, 9]:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X, y)
    pred = knn.predict(X_test_df)

    plt.figure(figsize=(10, 6))
    plt.scatter(X_test_df['embed_0'], X_test_df['embed_1'], c=pred, alpha=0.5)
    plt.xlabel('embed_0')
    plt.ylabel('embed_1')
    plt.title(f'Test Data Classification with k={k}')
    plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Load your dataset
df = pd.read_excel('training_mathbert 3 test.xlsx')

feature_1 = 'embed_0'  
feature_2 = 'embed_1'  
class_label = 'output'

X = df[[feature_1, feature_2]]
y = df[class_label]

# Generate test data
x_test = np.arange(X[feature_1].min(), X[feature_1].max(), 0.1)
y_test = np.arange(X[feature_2].min(), X[feature_2].max(), 0.1)
xx, yy = np.meshgrid(x_test, y_test)
X_test = np.c_[xx.ravel(), yy.ravel()]

# Convert X_test to DataFrame with the same feature names
X_test_df = pd.DataFrame(X_test, columns=[feature_1, feature_2])

# Define the parameter grid for k
param_grid = {'n_neighbors': np.arange(1, 20)}

# Initialize the GridSearchCV
grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, scoring='accuracy')

# Fit the model
grid_search.fit(X, y)

# Get the best k value
best_k = grid_search.best_params_['n_neighbors']
print(f'The best value of k is: {best_k}')

# Train and evaluate the model with the best k
knn_best = KNeighborsClassifier(n_neighbors=best_k)
knn_best.fit(X, y)
y_pred = knn_best.predict(X_test_df)

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X_test_df[feature_1], X_test_df[feature_2], c=y_pred, cmap='coolwarm', alpha=0.5)
plt.xlabel(feature_1)
plt.ylabel(feature_2)
plt.title(f'Test Data Classification with best k={best_k}')
plt.show()
```
